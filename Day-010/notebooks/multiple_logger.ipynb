```python
import logging
```

```python
#Two seperate loggers
logger1=logging.getLogger("logger1")
logger2=logging.getLogger("logger2")

#Setting different logging levels
logger1.setLevel(level=logging.DEBUG)
logger2.setLevel(level=logging.ERROR)

#Creating a console handler
stream=logging.StreamHandler()
logger1.addHandler(stream)
logger2.addHandler(stream)
```

```python
logger1.debug("This is a debug message from logger1.")
logger2.error("This is an error message from logger2.")
```
