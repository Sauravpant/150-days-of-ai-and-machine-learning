### Introduction to Seaborn
#### Seaborn is a Python based visualization library based on Matplotlib that provides a high level interface for drawing interactive and attractive statistical graphics. It helps in creating complex visualizations with just few lines of code.

### Importing Seaborn

```python
import seaborn as sns
```

### Basic Plotting with Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt

tips=sns.load_dataset("tips")
tips
```

### Scatter plot

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

### Line Plot

```python
sns.lineplot(data=tips,x='size',y='total_bill')
plt.title('Line Plot of Total bill by size')
```

### Bar Plot

```python
sns.barplot(data=tips,x='day',y='total_bill')
plt.title("Bar plot of Total bill")
```

### Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=tips,color='red')
```

### Violin plot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

### Histogram

```python
sns.histplot(data=tips,x='total_bill',bins=10,kde=True,color='green')
```

### Pair plot

```python
sns.pairplot(tips)
```

### KDE plot

```python
sns.kdeplot(x='total_bill',data=tips,fill=True,color='blue')
```

```python
corr=tips[['total_bill','tip','size']].corr()
corr
```

### Heatmap

```python
sns.heatmap(corr,annot=True,color='red')
```
