# Introduction to Matplotlib

### Matplotlib is a powerful and versatile open-source plotting library for Python, designed to visualize data in a variety of formats.

### pyplot API
#### matplotlib.pyplot is a state based interface that is mainly intended for interactive use and simple script-based plot generation

### Importing Matplotlib

```python
import matplotlib.pyplot as plt
```

### Figures Vs Axes in Matplotlib
#### Figure is the entire window that can contain multiple plots and axes is the actual plotting area that has x and y axes.

### Core Plot Types in Matplotlib
#### Line Plot -> Shows trend over continious data
#### Scatter plot -> Shows relationship between 2 variables
#### Bar Plot -> Compare variables
#### Histogram -> Shows distribution of single variable

### Line Plot plot(x,y)

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot')
plt.show()
```

### Scantter Plot scatter(x,y)

```python
plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot")
plt.show()
```

### Bar Plot bar(x,height)

```python
categories = ["A", "B", "C", "D"]
values = [10, 24, 36, 18]

plt.bar(categories, values)
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Bar Plot")
plt.show()
```

### Histogram hist(x)

```python
data = [22, 21, 20, 23, 24, 25, 22, 21, 26, 27, 24, 23]

plt.hist(data, bins=5)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

### Plot Structure

#### plt.figure() 

```python
plt.figure()
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

#### plt.subplots()

```python
fig, ax = plt.subplots(1,2)
ax[0].plot([1,2,3], [4,5,6])   # left plot
ax[1].plot([1,2,3], [6,5,4])   # right plot
plt.show()
```

### Labels, Titles and Legends

```python
x = [1,2,3,4]
y = [2,4,6,8]
```

#### X-Axis and Y-Axis Labels

```python
plt.plot(x, y)
plt.xlabel("Time (days)")   # X-axis label
plt.ylabel("Value")         # Y-axis label
plt.show()
```

#### Title

```python
plt.plot(x, y)
plt.xlabel("Time (days)")
plt.ylabel("Value")
plt.title("Growth over Time")  # Plot title
plt.show()
```

#### Legend : Shows the label for plots if there are multiple plots

```python
plt.plot(x, y, label="Line A")
plt.plot(x, [1,3,5,7], label="Line B")
plt.xlabel("Time (days)")
plt.ylabel("Value")
plt.title("Multiple Lines With Legend")
plt.legend()  # Display legend
plt.show()
```

### Basic Styling

```python
#### linestyle : Shows how the line looks 
```

```python
plt.plot([1,2,3,4], [2,4,6,8], linestyle="--")  # dashed line
plt.plot([1,2,3,4], [1,3,5,7], linestyle=":")   # dotted line
plt.show()
```

#### Marker: Shows points on the line

```python
plt.plot([1,2,3,4], [2,4,6,8], marker="o")  # circles
plt.plot([1,2,3,4], [1,3,5,7], marker="s")  # squares
plt.show()
```

#### Line Width : Makes line thicker/thinner

```python
plt.plot([1,2,3,4], [2,4,6,8], linewidth=3)  # thicker
plt.plot([1,2,3,4], [1,3,5,7], linewidth=0.5)  # thinner
plt.show()
```

#### Color: Change colors of line/bars

```python
plt.plot([1,2,3,4], [2,4,6,8], color="red")
plt.plot([1,2,3,4], [1,3,5,7], color="#00FF00")  # hex code
plt.show()
```

#### Aipha: Used to control the transparency of line

```python
plt.plot([1,2,3,4], [2,4,6,8], alpha=0.5)  # 50% transparent
plt.plot([1,2,3,4], [1,3,5,7], alpha=1.0)  # fully opaque
plt.show()
```

```python
plt.plot([1,2,3,4], [2,4,6,8], color="red", linestyle="--", marker="o", linewidth=2, alpha=0.7)
plt.plot([1,2,3,4], [1,3,5,7], color="blue", linestyle="-.", marker="s", linewidth=1, alpha=1)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Styled Lines Example")
plt.legend(["Red Line", "Blue Line"])
plt.show()
```
