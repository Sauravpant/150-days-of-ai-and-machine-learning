## SQLite

### SQLite is a self contained, serverless and zero configuration database engine that is widely used for embedded database systems. Unlike other relational databases like MySQL and Postgres SQLite runs inside our appllication and all read/write operatiions happens inside a single .db file . It only allows a single writer at a time.

### SQLite3 comes inbuilt with python

```python
import sqlite3
```

```python
connection=sqlite3.connect('database.db')
connection
```

#### To interact with the database and execute sql queries we need to make a cursor object

```python
cursor = connection.cursor()
```

```python
# Create a table
cursor.execute(''' 
CREATE TABLE IF NOT EXISTS 
    users (id INTEGER PRIMARY KEY AUTOINCREMENT, 
    name TEXT, 
    age INTEGER,
    department TEXT
    )
''')
#By default autocommit is off in SQLite, so we need to commit the changes
connection.commit()
```

```python
cursor.execute(' SELECT * FROM users ')
```

```python
#Insert the data
cursor.execute('''
    INSERT INTO users (name, age, department)
    VALUES ('Saurav', 20, 'Computer Science'),
           ('Swastika', 19, 'Biotechnology'),
           ('Gita', 28, 'Marketing')
''')
connection.commit()
```

```python
data=cursor.execute(' SELECT * FROM users ')
rows=data.fetchall()
rows
```

```python
print("Users in the database:")
for row in rows:
  print(f"ID: {row[0]}, Name: {row[1]}, Age: {row[2]}, Department: {row[3]}")
```

```python
# Updating data
cursor.execute('UPDATE USERS SET age=20 WHERE name="Swastika"')
connection.commit()
```

### CRUD Operations with Students data

```python
connection = sqlite3.connect('students.db')
cursor = connection.cursor()

# Create students table
cursor.execute('''
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    age INTEGER,
    grade TEXT,
    department TEXT
)
''')

connection.commit()
```

### Handling Transactions in SQLite3

```python
def insert_mutiple_students(student1,student2):
  try:
    cursor.execute('INSERT INTO students VALUES (?,?,?,?,?)',student1)
    cursor.execute('INSERT INTO students VALUES (?,?,?,?,?)',student2)
    connection.commit() #Commit the changes
    print('Insertion successful')
  except Exception as e:
    print('An error occured',e)
    connection.rollback() #Rollback in case of error
  

insert_mutiple_students( (100,'Zara',20,'A','Computer Science'), (200,'Zayn',21,'B','Biotechnology') )
```

### Backing Up and Restoring Data

```python
import shutil
def backup_database(source_db, backup_db):
    try:
        shutil.copy(source_db, backup_db)
        print(f'Backup of {source_db} created as {backup_db}')
    except Exception as e:
        print('An error occurred during backup:', e)
        
backup_database('students.db','students_backup.db')
```

```python
def restore_database(backup_db, source_db):
    try:
        shutil.copy(backup_db, source_db)
        print(f'Database {source_db} restored from {backup_db}')
    except Exception as e:
        print('An error occurred during restoration:', e)
        
restore_database('students_backup.db', 'students.db')
```
