```python
import matplotlib.pyplot as plt
import numpy as np
```

### Multiple Subplots


```python
fig,ax=plt.subplots(2,2)
ax[0,0].plot([1,2,3], [1,4,9])
ax[0,1].scatter([1,2,3], [3,2,1])
ax[1,0].bar(["A","B","C"], [5,7,3],color=["blue","yellow","pink"],width=0.6)
ax[1,1].hist([1,2,2,3,3,3,4])
```

```python
fig, ax = plt.subplots(1, 2,sharey=True)
ax[0].plot([1, 2, 3], [1, 4, 9])
ax[1].plot([1, 2, 3], [1, 2, 3])

ax[0].set_ylabel("Value")
ax[1].set_xlabel("X axis2")
ax[0].set_xlabel("X axis1")

plt.show()
```

```python
fig, ax = plt.subplots(2, 1,sharex=True)
ax[0].plot([1, 2, 3], [1, 4, 9])
ax[1].plot([1, 2, 3], [1, 2, 3])

ax[0].set_xlabel("Value")
ax[1].set_ylabel("X axis2")
ax[0].set_ylabel("X axis1")

plt.show()
```

```python
epochs=np.arange(0,50)
loss=np.random.rand(50)*0.01+0.9
```

```python
fig, ax=plt.subplots(1,2)
ax[0].plot(epochs,loss,color='blue',linewidth=2,linestyle='-')
ax[1].scatter(epochs,loss,color="red",marker='x')
ax[0].set_xlim(0,50)
ax[1].set_xlim(0,50)
ax[1].set_ylim(0.8,1.2)
ax[0].set_ylim(0.8,1.2)
ax[0].set_xlabel("Epochs")
ax[0].set_ylabel("Loss")
ax[0].set_title("Training Loss over Epochs")
```

### Pie Chart
