### Handling Outliers

```python
import numpy as np
import seaborn as sns
```

```python
marks=[-100,-200,45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74,150,170,180]
minimum,Q1,median,Q3,maximum=np.quantile(marks,[0,0.25,0.50,0.75,1.0])
```

```python
minimum,Q1,median,Q3,maximum
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
print(lower_bound,upper_bound)
```

```python
sns.boxplot(marks)
```

```python
def remove_outliers(marks):
    final_marks=[]
    for mark in marks:
        if mark>=lower_bound and mark<=upper_bound:
            final_marks.append(mark)  
    return final_marks
            
```

```python
print(f'Marks before outlier removal: {marks}')
```

```python
print(f'Marks after removing outliers: {remove_outliers(marks)}')
```
